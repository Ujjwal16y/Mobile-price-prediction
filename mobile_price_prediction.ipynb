```python
y=train['price_range']
x=train.drop('price_range',axis=1)


```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
xtrain=ss.fit_transform(xtrain)
xtest=ss.transform(xtest)
```

```python
xtrain
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
pred=lr.predict(xtest)
```

```python
print("accuracy score on testing:",accuracy_score(ytest,pred))
print('accuracy score on training:-',(accuracy_score(lr.predict(xtrain),ytrain)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
pred=dtc.predict(xtest)
print("accuracy score on tets:",accuracy_score(ytest,pred))
print("accuracy score:",accuracy_score(ytest,pred))
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(xtrain,ytrain)
pred=sv.predict(xtest)
print("accuracy score:",accuracy_score(ytest,pred))
```
